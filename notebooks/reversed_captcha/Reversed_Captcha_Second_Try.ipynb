```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium import webdriver
import base64
import requests

key_word = "shark"

def open_browser_for_search():
    chrome_options = Options()
    chrome_options.add_experimental_option("detach", True)
    
    #Opens up web driver and goes to Google Images
    driver = webdriver.Chrome('C:/Driver/chromedriver.exe', options=chrome_options)
    driver.get('https://images.google.com/imghp?hl=nl&gl=nl')
    return driver

def close_cookie_popup(driver):
    cookie_popup = driver.find_element(By.XPATH, '//*[@id="L2AGLb"]/div')
    cookie_popup.click()
    
def search_key_word(term_to_search, driver):
    box = driver.find_element(By.CLASS_NAME, "gLFyf")
    box.send_keys(term_to_search)
    box.send_keys(Keys.ENTER)
    
def save_my_image(base64string, path):
    base64string = strip_base64_info_from_string(base64string)
    my_str_as_bytes = str.encode(base64string)
    with open(path, "wb") as fh:
        fh.write(base64.decodebytes(my_str_as_bytes))
        
def strip_base64_info_from_string(base64string):
    splitted_string = base64string.split(",")
    return splitted_string[1] # right side of the ,
    
def strip_base64_info_from_string(base64string):
    splitted_string = base64string.split(",")
    return splitted_string[1] # right side of the ,

def save_image_based_on_url(url, path):
    with open(path, "wb") as f:
        f.write(requests.get(url).content)
    
def save_nth_image(n, path, driver):
    try:
        driver.find_element(By.XPATH, '//*[@id="islrg"]/div[1]/div[' + str(n) + ']/a[1]/div[1]/img').click()
        image_base64 = driver.find_element(By.XPATH, '//*[@id="Sva75c"]/div/div/div[3]/div[2]/c-wiz/div/div[1]/div[1]/div[3]/div/a/img').get_attribute('src')
        if image_base64.startswith("http"):
            save_image_based_on_url(image_base64, path)
        else:
            save_my_image(image_base64, path)
    except Exception as e:
        print(e)
        print("Saving " + path + " went wrong :-( ")
        
def get_n_images_by_label(label_name, n):
    root_directory = "C:\ReversedCaptcha\Images\\"
    create_directory_for_label(root_directory, label_name)
    
    driver = open_browser_for_search()
    close_cookie_popup(driver)
    search_key_word(label_name, driver)
    
    for i in range(1, n+1):
        path = root_directory + label_name + "\\" + label_name + str(i) + ".jpg"
        save_nth_image(i, path, driver)
    driver.quit()

# Get directory and 10 images for 1 label
# get_n_images_by_label("shark", 10)
```

# stuur per label deze 10 plaatjes op

We hebben nu een directory met 10 plaatjes per categorie (10*1000). Nu gaan we deze 1 voor 1 naar de API sturen om de responses op te halen.

```python
# This is how the response will be stored in the persistent storage
class LabelEntity:
    label: str
    status: str
    result: str
    def EntityToJson(self):
        return '{"status": "' + self.status + '", "result": "' + self.result + '", "label": "' + self.label + '"}'
    
def create_json_file_for_label(label_name, json_string):
    os.mkdir("C:\ReversedCaptcha\Images\LABEL_NAME\Result".replace("LABEL_NAME", label_name))
    path = "C:\ReversedCaptcha\itemname\Result\LABEL_NAME.json"
    path = path.replace("LABEL_NAME", label_name)

    with open(path, 'w') as f:
        json.dump(json_string, f)
        
create_json_file_for_label("affenpinscher")
    
    
```

```python
# one
def send_image_to_api(image_label, image_n):
    url = "https://europe-west1-coding-challenge-platform.cloudfunctions.net/ctf-reverse-captcha-5736741495373824"

    image_path = "C:\ReversedCaptcha\itemname\itemnamenumberofimage.png".replace("itemname", image_name)
    image_path = image_path.replace("numberofimage", str(image_n))
    
    with open(image_path, "rb") as file:
        files = {'file': file}
        
        response = requests.post(url, headers={"Authorization":"{5736741495373824}"}, files=files)
        response_as_string = str(response.json())
        response_as_string = response_as_string.replace("'", '"')
        
    return response_as_string

# multiple
def send_label_images_to_api(image_label):
    n_images_per_label = 10
    
    
    json_string = '{ "values": [' # standard content of the JSON file
    for x in range(0, n_images_per_label):
        result = send_image_to_api(image_label, x+1) # +1 because images range starts at 1
        
        # format JSON file to dump
        json_string += result
        if x != n_images_per_label-1:
            json_string += "," # add , for next item in json array
    
    json_string += ']  }'
    create_json_file_for_label(image_label, json_string)
        
    
```
