```python
def open_browser_for_search():
    chrome_options = Options()
    chrome_options.add_experimental_option("detach", True)
    
    #Opens up web driver and goes to Google Images
    driver = webdriver.Chrome('C:/Driver/chromedriver.exe', options=chrome_options)
    driver.get('https://images.google.com/imghp?hl=nl&gl=nl')
    return driver
```

```python
def close_cookie_popup(driver):
    cookie_popup = driver.find_element(By.XPATH, '//*[@id="L2AGLb"]/div')
    cookie_popup.click()  
```

```python
def search_key_word(term_to_search, driver):
    box = driver.find_element(By.CLASS_NAME, "gLFyf")
    box.send_keys(term_to_search)
    box.send_keys(Keys.ENTER)
```

```python
# function for saving a base64 image to the directory

def save_my_image(base64string, path):
    base64string = strip_base64_info_from_string(base64string)
    my_str_as_bytes = str.encode(base64string)
    with open(path, "wb") as fh:
        fh.write(base64.decodebytes(my_str_as_bytes))

def strip_base64_info_from_string(base64string):
    splitted_string = base64string.split(",")
    return splitted_string[1] # right side of the ,
```

```python
import requests
import json
```

```python
# Example of what the response looks like
response = "{'status': 'ok', 'result': '0.000124692'}"
response = response.replace("'", '"')
file_name = "shark"
```

```python
class LabelEntity:
    label: str
    status: str
    result: str
    def EntityToJson(self):
        return '{"status": "' + self.status + '", "result": "' + self.result + '", "label": "' + self.label + '"}'

def response_to_label_entity(json_as_string_response, file_name):

    response_as_json = json.loads(json_as_string_response)

    label_entity = LabelEntity()
    label_entity.label = file_name
    label_entity.result = response_as_json['result']
    label_entity.status = response_as_json['status']

    return label_entity

def save_label_entity_to_file(label_entity):
    with open('reversed_captcha_responses.json','r+') as f:
        data = json.load(f)
        le_as_json = label_entity.EntityToJson()
        data['values'].append(json.loads(le_as_json))
        # Sets file's current position at offset.
        f.seek(0)
        # convert back to json.
        json.dump(data, f, indent = 4)
        #print(data['values'])
        print("added " + label_entity.label)
```

```python
def send_api_request(image_name):
    url = "https://europe-west1-coding-challenge-platform.cloudfunctions.net/ctf-reverse-captcha-5736741495373824"

    image_path = "C:\Driver\Images\itemname.png".replace("itemname", image_name)
    
    with open(image_path, "rb") as file:
        files = {'file': file}
        #print(files)
        response = requests.post(url, headers={"Authorization":"{5736741495373824}"}, files=files)
        response_as_string = str(response.json())
        response_as_string = response_as_string.replace("'", '"')
        #print(response_as_string)
        response_to_save = response_to_label_entity(response_as_string, image_name)
        #print(response_to_save)
        save_label_entity_to_file(response_to_save)

def send_api_requests():
    labels = get_all_labels()
    
    not_found_images = []
    for label in labels:
        try:
            send_api_request(label)
        except:
            not_found_images.append(label)
            
    print("images that are not found: " + str(not_found_images))
        
```
